# Introduction to Matplotlib

Matplotlib is a plotting library for Python. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK.

## Key Concepts

- **Figure**: The whole figure which may contain one or more axes.
- **Axes**: The area that we plot on. A figure can contain multiple axes.
- **Axis**: Number-line like objects and take care of generating the graph limits.
- **Artist**: Everything you can see on the figure is an artist, like Text objects, Line2D objects, etc.

## Basic Syntax

Here is a simple example:

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [1, 4, 9])
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Simple Plot')
plt.show()
```

In this example, `plt.plot()` is used to plot data points. The `xlabel` and `ylabel` functions are used to set labels for the x and y-axis, respectively. The `title` function is used to set the title of the plot. Finally, `show()` is used to display the plot.

## Line Plot

### Syntax

```python
plt.plot(x, y, [fmt], **kwargs)
```

- `x`: The x-coordinates of the points.
- `y`: The y-coordinates of the points.
- `fmt`: A format string to specify color, marker, and line type.
- `**kwargs`: Additional keyword arguments to customize the plot.

### Example

Let's create a simple line plot.

````python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
````

## Bar Plot

### Syntax

```python
plt.bar(x, height, width=0.8, bottom=None, align='center', **kwargs)
```

- `x`: The x coordinates of the bars.
- `height`: The height of the bars.
- `width`: The width of the bars.
- `bottom`: The y-coordinate of the bottom of the bars.
- `align`: Alignment of the bars ('center' or 'edge').
- `**kwargs`: Additional keyword arguments.

### Example

Let's create a simple bar plot.

````python
# Create data
x = ['Apple', 'Banana', 'Cherry', 'Date']
y = [5, 7, 3, 8]

# Create a bar plot
plt.bar(x, y)
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.title('Fruit Quantity')
plt.show()
````
